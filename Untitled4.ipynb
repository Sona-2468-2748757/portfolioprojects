```python
from scipy.stats import chi2_contingency

categorical_vars = ['gender', 'received_promotions', 'referred_by_friend']

for col in categorical_vars:
    contingency = pd.crosstab(df[col], df['is_churned'])
    
    # Skip if contingency table is empty or constant
    if contingency.shape[0] < 2 or contingency.shape[1] < 2:
        print(f"\nSkipping {col}  not enough variation for chi-square test.")
        continue
    
    chi2, p, dof, expected = chi2_contingency(contingency)
    print(f"\nChi-square test for {col}")
    print(f"p-value = {p:.4f}")
    if p < 0.05:
        print(" Significant relationship with churn.")
    else:
        print("No significant relationship with churn.")
```

```python
from scipy.stats import ttest_ind

churned = df[df['is_churned'] == 1]['average_watch_hours']
retained = df[df['is_churned'] == 0]['average_watch_hours']

t_stat, p_value = ttest_ind(churned, retained, equal_var=False, nan_policy='omit')
print(f"T-test p-value: {p_value:.4f}")
if p_value < 0.05:
    print(" Average watch hours differ significantly between churned and retained users.")
else:
    print("No significant difference in average watch hours between churned and retained users.")
```
